```python
import pandas as pd
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movie_df = pd.read_csv('netflix_titles.csv')

movie_df.head()
```

```python
userInput = [{'title':'The Walking Dead','rating':3},
             {'title':'BoJack Horseman','rating':4},
             {'title':'Black Mirror','rating':4.5},
             {'title':'After Life','rating':3.5},
             {'title':'Community','rating':3.5}] 


userMovies = pd.DataFrame(userInput)
userMovies
```

```python
j_movie_df = movie_df.copy()
movie_df['listed_in'] = movie_df.listed_in.str.split(', ')
movie_df.head()
```

```python
 _movie_df = movie_df.copy()


for index, row in movie_df.iterrows():
    for category in row['listed_in']:
        _movie_df.at[index, category] = 1 
_movie_df = _movie_df.drop(columns=['director','listed_in','type','rating','cast','country','date_added','release_year','duration','description'])        
_movie_df = _movie_df.fillna(0)  
_movie_df['show_id'] = _movie_df['show_id'].map(lambda x : x.lstrip('s'))
_movie_df = _movie_df.set_index(_movie_df['show_id'])
_movie_df.head()
```

```python
j_movie_df = j_movie_df.drop(columns = ['type','director','cast','director','cast','country','date_added','release_year','rating','duration','listed_in','description'])
j_movie_df['show_id'] = j_movie_df['show_id'].map(lambda x : x.lstrip('s'))
j_movie_df.head()
```

```python
inputMatch = j_movie_df[j_movie_df['title'].isin(userMovies['title'].tolist())]
userMovies = pd.merge(inputMatch, userMovies)
userMovies.head()
```

```python
inputMatch = _movie_df[_movie_df['show_id'].isin(userMovies['show_id'].tolist())]
inputMatch
```

```python
userMovies = userMovies.reset_index(drop=True)
```

```python
inputMatch = inputMatch.reset_index(drop=True)
ord_table = inputMatch.drop(columns = ['title','show_id'])
ord_table.head()
```

```python
user = ord_table.transpose().dot(userMovies['rating'])
user
```

```python
u_ord_table = _movie_df.set_index(_movie_df['title'])
u_ord_table = u_ord_table.drop(columns = ['title','show_id'])
u_ord_table.head()
```

```python
recommendation_df = ((u_ord_table*user).sum(axis=1))/(user.sum())
j_movie_df.loc[j_movie_df['show_id'].isin(recommendation_df.head(20).keys())].drop(columns = ['show_id'])
```
